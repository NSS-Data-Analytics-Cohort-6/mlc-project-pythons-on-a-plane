```python
import json
import pandas as pd
```

```python
# Opening JSON file
g = open('C:/Users/iulia/Documents/NSS/projects/mlc-project-pythons-on-a-plane/data/genres.json')
 
# returns JSON object as a dictionary
genres_json = json.load(g)

# since the value lists have varying lengths, we have to use a lambda function to iterate over the keys and values as a series
genres = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in genres_json.items()]))
genres.head(20)
```

```python
#switch rows with columns (transpose)
genres = genres.T
```

```python
#subsetting and splitting values in the
genres_fix = genres[0].str.split(',', expand = True)
genres_fix.head()
```

```python
# axis =1 tells drop the 1st column
genres = genres.drop(0, axis = 1)
genres.head()
```

```python
genres_final = pd.merge(genres, genres_fix, left_index = True, right_index = True, how = 'outer')
pd.set_option('max_columns', None)
genres_final.head()
```

```python
genres_final['1_y'].head()
```

```python
genres_finalx2 = genres_final.T
genres_finalx2.head()
```

```python
genres_melt = pd.melt(genres_finalx2)
genres_melt.head(60)
```

```python
genres_melt.columns = ['Artist','Genre']
```

```python
genres_melt['Genre'] = genres_melt['Genre'].str.upper()
```

```python
genres_melt = genres_melt.dropna()
genres_melt
```
