```python
import json
import pandas as pd
```

```python
# opening json file
g = open('../data/genres.json')

# returns JSON object as a dictionary
genres_json = json.load(g)

# since the value lists have varying lengths, we have to use a lambda function to iterate over the keys and values as a series
genres = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in genres_json.items()]))
```

```python
# switch rows & columns
genres = genres.transpose()

# drop null values
#genres = genres.dropna()

#switch artists to rows / genres to columns
#genres = pd.melt(genres)

# reset index
genres = genres.reset_index(drop=False)

# rename columns
genres.columns = ['artist', 'genre1', 'genre2', 'genre3', 'genre4', 'genre5', 'genre6', 'genre7', 'genre8', 'genre9', 'genre10', 'genre11', 'genre12', 'genre12']

genres.head(20)
```

```python
# create subset with genre1 comma values
# https://stackoverflow.com/questions/11350770/filter-pandas-dataframe-by-substring-criteria
genre_commas = genres[genres['genre1'].str.contains(", ")]

# drop columns with all NaN values
# https://stackoverflow.com/questions/45147100/pandas-drop-columns-with-all-nans
genre_commas = genre_commas.dropna(axis=1, how='all')

genre_commas.head(50)
```
